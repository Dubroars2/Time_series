# Рассмотрим датасет International airline passengers: monthly totals in thousands

```python
#Тест на стационарность 
def test_stationarity(timeseries):#передаем временной ряд (массив)
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')# autolag коэффициент при каких значениях наиболее правдоподобные результаты
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    #в результате получаем значения статисики, пи велью, какой лаг (какой сдвиг делать) и кол-во наблюдений
    for [key, value] in dftest[4].items(): # перебираем критичные значени
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
#Визуализация автокорреляции и частичной автокорреляции
def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    test_stationarity(y)#тест на стационарность ряда
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))
#параметры графика
        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)#автокорреляция
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)#частичная

        plt.tight_layout()
    return
```

# Будем использовать метод Хольт-Винтерс

Используем этот метод, так как нам нужно учитывать сезонность 

```python
#инициализируем наш тренд 

def initial_trend(series, season_len):#временной ряд и season_len длительность сезона, которую мы сами обозначили 
    return sum([float(series[i + season_len] - series[i]) / season_len]) / season_len
#проходим по длительности сезона
```

```python
#инициализация сезонных компенентов
def initial_seasonal_components(series, slen): #подаем вр ряд и длину сезона
    seasonals = {} #будем хранить сезоны в словаре
    season_averages = [] #ср значение по сезонам
    n_seasons = int(len(series)/slen) #кол-во полных сезонов, которые есть
    #берем кол-во значений временного ряда и делим на длительность сезона
    # compute season averages
    for j in range(n_seasons): #перебираем каждый сезон по отдельности
        season_averages.append(sum(series[slen*j:slen*j+slen])/float(slen))
        #длина сезона умноженная на номер сезона и делим на размер сезона (12 к примеру)
        #в первом случае будет от 0-12, потом от 12-24 и тд
    # compute initial values
    #посчитаем первые сезонные значения 
    for i in range(slen): #перебираем от 1 до размера сезонности
        sum_of_vals_over_avg = 0.0
        for j in range(n_seasons): # перебираем по кол-ву сезонов, которые есть
            sum_of_vals_over_avg += series[slen*j+i]-season_averages[j]
            #получаем разницу между текущим значением и средним сезон значением
        seasonals[i] = sum_of_vals_over_avg/n_seasons # усредняем среднее значение 
    return seasonals
```

```python
def triple_exponential_smoothing(series, slen, alpha, beta, gamma, n_preds):
    result = []#пока пустое
    seasonals = initial_seasonal_components(series, slen)
    for i in range(len(series)+n_preds):
        if i == 0: # initial values
            smooth = series[0]
            trend = initial_trend(series, slen)
            result.append(series[0])
            continue
        if i >= len(series): # we are forecasting
            m = i - len(series) + 1
         
            result.append((smooth + m*trend) + seasonals[i%slen])
                     
        else:
            val = series[i]
            last_smooth, smooth = smooth, alpha*(val-seasonals[i%slen]) + (1-alpha)*(smooth+trend)
            
            trend = beta * (smooth-last_smooth) + (1-beta)*trend
            seasonals[i%slen] = gamma*(val-smooth) + (1-gamma)*seasonals[i%slen]
            
            result.append(smooth+trend+seasonals[i%slen])
               
    return result
```

```python
def plot_tema(alpha, beta, gamma, ser=df1, ser_to_plot=df1, n_preds=24):
    tema = triple_exponential_smoothing(ser, 6, alpha, beta, gamma, n_preds)
    with plt.style.context('bmh'):
        plt.figure(figsize=(14, 8))
        plt.plot(ser_to_plot, color='blue',label='original')
        plt.plot(tema, color='red', linewidth='4', label='TEMA')
        plt.title("alpha={}, beta={}, gamma={}".format(alpha, beta, gamma))
        plt.legend()
```

```python
from scipy.optimize import minimize
from sklearn.metrics import mean_squared_error
```

```python
series = df1
```

```python
train, test, val = series[:60], series[60:70], series[70:]
```

```python
#Описываем функцию мсе
def mse(X):
    alpha, beta, gamma = X
    #в качестве x берем кортеж из 3-х значений альфа, бета, гамма
    result = triple_exponential_smoothing(train, 6, alpha, beta, gamma, len(test))
    #результат это тройное экспоненциальное сглаживание
    #подаем обучающую часть трейн, 6 это размер сезона, len(test) размер валидационной выборки 
    predictions = result[-len(test):]
    #отбираем предсказания, последние компаненты, которые есть в валидационной выборке 
    error = mean_squared_error(predictions, test)
    #и считаем ошибку предсказаний и валидационной выборки тест
    return error
```

```python
#применяем метод оптимайз
opt = minimize(mse, x0=[0,0,0], method="L-BFGS-B", bounds = ((0, 1), (0, 1), (0, 1)))
#bounds границы четкие альфа гамма и бетта от 0 до 1 чтобы не вылезали за их пределы
```

```python
alpha_opt, beta_opt, gamma_opt = opt.x
#opt.x какие значения Х оказались адекватные 
print(opt)
```

```python
plot_tema(alpha_opt, beta_opt, gamma_opt, ser=train, ser_to_plot=series[:70], n_preds=len(test))
```

# Будем использовать скользящую среднюю, так как у нас ряд стационарный 

```python
# Для скользящей средней нужно сделать функцию

def moving_average(series, n): #На вход передаем данные в формате сериас и кол-во (для окна ниже)
    if not isinstance(series, pd.Series):#Если не Формат сериас, то делаем этот массив сериас 
        series = pd.Series(series)
    return series.rolling(n).mean() # Задаем некое окно и к нему применяем агрегирующую функцию  
```

```python
ser = moving_average(df2, 10) #применим скользящую. Возьмем датафрейм и интервал за 10 дней
with plt.style.context('bmh'):
    plt.figure(figsize=(14, 8))
    plt.plot(df2, color='blue',label='original')
    plt.plot(ser, color='red', linewidth='4', label='Moving Average')
    plt.legend()
```

```python
series2 = df2
```

```python
def predict(series2, N, n_pred):
    new_series = series2.copy()
    for _ in range(n_pred): #На каждом шаге предсказания. n_pred сколько хотим предсказать
        new_series = new_series.append(pd.Series([new_series[-N:].mean()]), ignore_index=True)
        #будем брать окно из последних Н значений new_series[-N:].mean временного ряда
        #И будем постоянно добавлять наш временной ряд 
        #Каждый раз шаг за шагом  добавляется значение и окно сдвигается вправо 
    return new_series
```

```python
series_pred = predict(df2, 10, 15)
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(14, 8))
    plt.plot(series_pred, color='green', linewidth='4', label='Pred')
    plt.plot(df2, color='blue', linewidth='4', label='Real')
    plt.legend()
```
